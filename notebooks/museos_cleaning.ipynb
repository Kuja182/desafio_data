Limpieza de datos de Museos

Empezamos llamando a la base de datos

```python
import pandas as pd
df_museos = pd.read_csv("../data/madrid/museos/201132-0-museos.csv",sep=";")
```

Vemos que problemas podemos tener de valores nulos y como solucionarlos.

```python
df_museos.info()
```

Como vemos hay muchas columnas que practicamente no ofrecen información o estan repetidas y nos disponemos a eliminarlas

```python
df_museos.drop(["PK"],axis = 1,inplace=True)
df_museos.drop("DESCRIPCION", axis = 1, inplace=True)
df_museos.drop("PLANTA", axis = 1, inplace=True)
df_museos.drop("PUERTA", axis = 1, inplace=True)
df_museos.drop("ESCALERAS", axis = 1, inplace=True)
df_museos.drop("ORIENTACION", axis = 1, inplace=True)
df_museos.drop("COD-BARRIO", axis = 1, inplace=True)
df_museos.drop("COD-DISTRITO", axis = 1, inplace=True)
df_museos.drop("FAX", axis = 1, inplace=True)
df_museos.drop("Unnamed: 32", axis = 1, inplace=True)
df_museos.drop("TIPO", axis = 1, inplace=True)
```

Ya teniendo bastante concretado, vamos a meternos en cada columna y vamos a tratar los valores nulos.

```python
df_museos["DESCRIPCION-ENTIDAD"] = df_museos["DESCRIPCION-ENTIDAD"].fillna("Descripción no disponible")
df_museos["EQUIPAMIENTO"] = df_museos["EQUIPAMIENTO"].fillna("Equipamiento no disponible")
df_museos["EMAIL"] = df_museos["EMAIL"].fillna("Email no disponible")
```

```python
accessibility_mapping = {
    0: 'No accesible',
    1: 'Accesible',
    2: 'Instalación parcialmente accesible para personas con movilidad reducida',
    3: 'Sin información sobre accesibilidad para personas con movilidad reducida',
    4: 'Lengua de signos',
    5: 'Señalización podotáctil',
    6: 'Bucle de inducción magnético'
}

def replace_accessibility(access_list):
    return [accessibility_mapping[num] for num in access_list]

df_museos['ACCESIBILIDAD'] = df_museos['ACCESIBILIDAD'].apply(replace_accessibility)
```
