```python
import requests
import pandas as pd
```

```python
def obtener_monumentos():
    url = "https://datos.madrid.es/egob/catalogo/300356-0-monumentos-ciudad-madrid.csv"
    
    try:
        response = requests.get(url)
        response.raise_for_status()  # Verifica si la solicitud fue exitosa

        # Si la solicitud es exitosa, retorna el contenido en formato CSV
        return response.content

    except requests.exceptions.RequestException as e:
        print("Error al obtener los monumentos:", e)
        return None
```

```python
contenido_csv = obtener_monumentos()

# Verificar si se obtuvo el contenido y guardarlo en un archivo CSV
if contenido_csv:
    with open("../data/madrid/Monumentos/monumentos.csv", "wb") as archivo_csv:
        archivo_csv.write(contenido_csv)
    print("Se ha guardado el archivo monumentos.csv")
```
